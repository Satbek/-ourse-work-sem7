```python
import numpy as np
import research
import view
import lib.gradients as grad
import lib.data as data
import matplotlib.pyplot as plt
import decompozition
import lib.mesurements
%matplotlib inline
```

```python
def poisson_noised(im):
    step = im.min()
    if (step < 0):
        im -= step
    noise_mask = np.random.poisson(im)
    if (step < 0):
        im += step
    return noise_mask
```

```python
def my_exp(x,y,w1,w2):
    if (w1 == 0):
        return np.exp( w2*1j*y)
    return np.exp(w1*1j*x + w2*1j*y)
```

```python
def calc_xyz(M,N, noised = False):
    x,y = data.get_plane(-np.pi, np.pi, -np.pi, np.pi, N)
    x_,y_  = np.meshgrid(range(-2**(M-1), 2**(M-1)), range(-2**(M-1), 2**(M-1)))
    z = np.zeros(2**M* 2**M).reshape(2**M,2**M)
    for i in range(-2**(M-1), 2**(M-1)):
        for j in range (-2**(M-1), 2**(M-1)):
            im = my_exp(x,y,i,j)
            specture = np.fft.fft2(im)
            grad_X, grad_Y = grad.fried_model_gradient(im)
            X_H, Y_H = grad.Hudgin_gradien_model(im)
            if (noised):
                #как зашумлять комплексные числа?
                grad_X = poisson_noised(grad_X.real) + 1j*poisson_noised(grad_X.imag)
                grad_Y = poisson_noised(grad_Y.real) + 1j*poisson_noised(grad_Y.imag)
                X_H = poisson_noised(X_H.real) + 1j*poisson_noised(X_H.imag)
                Y_H = poisson_noised(Y_H.real) + 1j*poisson_noised(Y_H.imag)
            LH, HL, HH = decompozition.analyze(grad_X, grad_Y, X_H, Y_H)
            res = decompozition.syntesis({0:np.array([[np.mean(im)*2**N]])}, LH, HL, HH,N)
            z[i,j] = np.abs(specture)[j,i] / np.abs(np.fft.fft2(res[N]))[j,i]
    return x_,y_,z
```

```python
#todo decorator
import matplotlib.ticker as plticker
def make_plot(M, N, noised = False, size = (30,30)):
    x,y,z = calc_xyz(M,N, noised)
    from mpl_toolkits.mplot3d import Axes3D
    fig = plt.figure(figsize=size)
    ax = fig.gca(projection='3d')ff
    ax.plot_wireframe(x, y, z)
    #ax.plot_wireframe(x, y, z)
    #loc = plticker.MultipleLocator(base=0.2)
    #ax.yaxis.set_major_locator(loc)
    plt.show()
```
