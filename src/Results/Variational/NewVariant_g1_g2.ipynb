```python
def get_g1(u, N1, N2):
    h1, h2 = 2 * np.pi / N1, 2 * np.pi / N2
    x, y = np.linspace(-np.pi, np.pi, N1), np.linspace(-np.pi, np.pi, N2)
    g1 = np.zeros((len(x),len(y)))
    for n in range(1, N1):
        for m in range(1, N2):
            g1[n,m] = (u(x[n], (y[m] + y[m - 1]) / 2) - u(x[n-1], (y[m] + y[m - 1]) / 2)) / h1
    for n in range(N1):
        g1[n,0] = g1[n,1]
    for m in range(N2):
        g1[0,m] = g1[1,m]    
    return g1

def get_g2(u, N1, N2):
    h1, h2 = 2 * np.pi / N1, 2 * np.pi / N2
    x, y = np.linspace(-np.pi, np.pi, N1), np.linspace(-np.pi, np.pi, N2)
    g2 = np.zeros((len(x),len(y)))
    for n in range(1, N1):
        for m in range(1, N2):
            g2[n,m] = (u((x[n] + x[n-1]) / 2, y[m]) - u((x[n] + x[n-1]) / 2, y[m - 1])) / h2
    for n in range(N1):
        g2[n,0] = g2[n,1]
    for m in range(N2):
        g2[0,m] = g2[1,m]    
    return g2
```

```python
def get_G(dim, h):
    diag1 = np.array([1 if i == j else 0 for i in range(dim) for j in range(dim)]).reshape(dim, dim)
    diag2 = np.array([-1 if i - j == -1 else 0 for i in range(dim) for j in range(dim)]).reshape(dim, dim)
    return (diag1 + diag2) / h

def get_B(dim, h):
    diag1 = np.array([1 if i == j else 0 for i in range(dim) for j in range(dim)]).reshape(dim, dim)
    diag2 = np.array([1 if i - j == 1 else 0 for i in range(dim) for j in range(dim)]).reshape(dim, dim)
    return (diag1 + diag2) / 2
```

# Получение F не через матрицы.
Попробуем вычислить $F$ не через матрицы, а явно.

По формулам:

$(g_1, \phi_x) = \frac{(g^1_{st} - g^1_{s+1,t} + g^1_{s,t+1} - g^1_{s+1,t+1})h_2}{2}$

$(g_2, \phi_y) = \frac{(g^2_{st} - g^2_{s,t+1} + g^2_{s+1,t} - g^2_{s+1, t+1})h_1}{2}$

$F = \frac{(g_1,\, \phi_x)\, + \,(g_2,\, \phi_y)}{h_1h_2} = \frac{g^1_{st} - g^1_{s+1,t} + g^1_{s,t+1} - g^1_{s+1,t+1}}{2h_1} + \frac{g^2_{st} - g^2_{s,t+1} + g^2_{s+1,t} - g^2_{s+1, t+1}}{2h_2}$

```python
def create_F(g1, g2, h1, h2):
    F = np.zeros(g1.shape)
    for s in range(F.shape[0] - 1):
        for t in range(F.shape[1] - 1):
            F1 = (g1[s,t] - g1[s + 1,t] + g1[s,t + 1] - g1[s + 1,t + 1]) / (2  * h1)
            F2 = (g2[s,t] - g2[s, t + 1] + g2[s + 1, t] - g2[s + 1, t + 1]) / (2 * h2)
            F[s, t] = F1 + F2
    return F    
```

```python
import numpy as np
import sys
sys.path.append("../..")
import lib
from lib.data import super_gauss
from lib.mesurements import mse
from lib.data import normalize
from lib.variational_method.variational import method as met
import lib.variational_method.eigenvalues as eigenvalues

def _f1(g1, h):
    """
    F1 составлющая матрицы F
    :param g1: 2d array. Градиент по x
    :param h1: шаг сетки по x
    :param h2: шаг сетки по y
    :return: 2d array
    """
    dim = g1.shape
    tmp = np.dot(get_G(dim[0], h), g1)
    return np.dot(tmp, get_B(dim[1], h))

def _f2(g2, h):
    """
    F2 составляющая матрицы F
    :param g2: 2d array. Градиент по y
    :param h1: шаг сетки по x
    :param h2: шаг сетки по y
    :return: 2d array
    """
    dim = g2.shape
    tmp = np.dot(get_B(dim[0], h), g2)
    return np.dot(tmp, get_G(dim[1], h))

def _get_f_matrix(g1, g2, h1, h2):
    """
    Матрица F. Правая часть в разностной схеме метода.
    :param g1: 2d array. Градиент по x
    :param g2: 2d array. Градиент по y
    :param h1: Шаг сетки по x
    :param h2: Шаг сетки по y
    :return: 2d array
    """
    return _f1(g1, h1) + _f2(g2, h2)


def method(g1, g2, h1, h2, alpha, gamma):
    """
    Метод. Принимает градиенты по x, y. Параметры регуляризации alga, gamma.
    Возвращает восстановленный волновой фронт.
    :param g1: 2d array. Градиент по x
    :param g2: 2d array. Градиент по y
    :param h1: Шаг сетки по x
    :param h2: Шаг сетки по y
    :param alpha: параметр регуляризации
    :param gamma: параметр регуляризации
    :return:
    """
    #f = np.fft.fft2(_get_f_matrix(g1, g2, h1, h2))
    f = np.fft.fft2(create_F(g1, g2, h1, h2))
    
    lambda1 = eigenvalues.get_lambda(f.shape[0], h1)
    lambda2 = eigenvalues.get_lambda(f.shape[1], h2)

    mu1 = eigenvalues.get_mu(f.shape[0], h1)
    mu2 = eigenvalues.get_mu(f.shape[1], h2)

    res = np.zeros(f.shape, dtype=complex)
    for k in range(res.shape[0]):
        for l in range(res.shape[1]):
            res[k][l] = (lambda1[k] * mu2[l] + mu1[k] * lambda2[l] +
                         alpha * mu1[k] * mu2[l] + gamma * lambda1[k] * lambda2[l])
    res = np.fft.ifft2(f / res)
    return np.real(res)
```
